```python
#let us drop drop columns ['kepid', rowid', 'kepler_name'], as kepoi_name serves our purpose and has no nulls
# also drop 'koi_teq_err1', & koi_teq_err2 columns having all nulls
# also drop 'koi_tci_delivname', 'koi_tce_plnt_num' which are only for housekeeping conventions
# also drop koi_disposition, which is from archive and not on the basis of kepler data
#'ra', and 'dec'are in degrees, shall we convert them to radians?  (not really, because it would be a linear conversion only)

df = df.drop(['kepid', 'rowid', 'kepler_name', 'koi_disposition','koi_teq_err1', 'koi_teq_err2', 'koi_tce_plnt_num','koi_tce_delivname'], axis =1)
df = df.drop([])
df.set_index('kepoi_name', drop =True, inplace = True)
df.head()
```

<b>koi_score value needs some change:</b>
Its a value between 0 and 1 that indicates the confidence in the KOI disposition. For CANDIDATEs, a higher value indicates more confidence in its disposition, while for FALSE POSITIVEs, a higher value indicates less confidence in that disposition. 

It means that to avoid confusing our algorithm to determine koi_pdisposition,  we should keep the value high for disposition confidence. i.e. subtract this value from 1 for FALSE POSITIVES.

```python
df['koi_score'] = np.where(df.koi_pdisposition == 'FALSE POSITIVE', 1-df['koi_score'], df['koi_score'])
df.head()
```

#### Take note that all err1 & err2 related features are uncertainities associated with main feature and therefore no relation to target variable.
#Hence we should either remove these columns, or better we should add average of +ve and -ve uncertainities to the main feature.
#Also this should be done before feature scaling step, as errors are defined on the scale of main feature and should not be independently scaled

```python
df.columns
```

```python
# adding average of +/-ve uncertainity error to main feature
df['koi_period'] = df['koi_period']+(df['koi_period_err1']+df['koi_period_err1'])/2
df['koi_time0bk'] = df['koi_time0bk']+(df['koi_time0bk_err1']+df['koi_time0bk_err2'])/2
df['koi_impact'] = df['koi_impact']+(df['koi_impact_err1']+df['koi_impact_err2'])/2
df['koi_duration'] = df['koi_duration']+(df['koi_duration_err1']+df['koi_duration_err2'])/2
df['koi_depth'] = df['koi_depth']+(df['koi_depth_err1']+df['koi_depth_err2'])/2
df['koi_prad'] = df['koi_prad']+(df['koi_prad_err1']+df['koi_prad_err1'])/2
df['koi_insol'] = df['koi_insol']+(df['koi_insol_err1']+df['koi_insol_err2'])/2
df['koi_steff'] = df['koi_steff']+(df['koi_steff_err1']+df['koi_steff_err2'])/2
df['koi_slogg'] = df['koi_slogg']+(df['koi_slogg_err1']+df['koi_slogg_err2'])/2
df['koi_srad'] = df['koi_srad']+(df['koi_srad_err1']+df['koi_srad_err2'])/2

# now dropping err1 & err2 columns
df = df.drop(['koi_period_err1','koi_period_err2', 
              'koi_time0bk_err1','koi_time0bk_err2', 
              'koi_impact_err1', 'koi_impact_err2',
       'koi_duration_err1', 'koi_duration_err2', 
       'koi_depth_err1', 'koi_depth_err2', 'koi_prad_err1',
       'koi_prad_err2', 'koi_teq', 'koi_insol', 'koi_insol_err1',
       'koi_insol_err2', 'koi_model_snr', 'koi_steff', 'koi_steff_err1',
       'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2',
       'koi_srad', 'koi_srad_err1', 'koi_srad_err2'], axis=1)

```

```python
# Import Modules
from sklearn import model_selection
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier


from datetime import datetime
```

```python
%%time

# feature & target data for training our models
X = X_scaled
y = y_categorical

# classification algorithms to be used (used all default valuse, except a few one where future warnings were issued)
classifiers = [
    SGDClassifier(max_iter=1000, tol=1e-3),
    LinearSVC(dual = False),
    KNeighborsClassifier(),
    SVC(gamma = 'auto'),
    RandomForestClassifier(n_estimators = 100, max_depth = 50),
    MLPClassifier(max_iter = 1000) # default for max-iter is 200, but its failing to converge with defaults value
    ]

models = ['SGD', 'LinearSVC', 'KNN', 'SVC',
         'RandomForest', 'MLP']



# evaluate each model in sequence

results = []
names = []
scoring = 'accuracy'
for model, clf in zip(models, classifiers):
    start_time = datetime.now()
    kf = model_selection.KFold(n_splits=3, random_state=11) # using 3 Kfolds
    cv_results = model_selection.cross_val_score(clf, X, y, cv=kf, scoring=scoring)
    results.append(cv_results)
    names.append(model)
    execution_time = datetime.now()- start_time
    
    print(f"Model {model} cross validation results with 3 folds: {cv_results}.")
    print(f"Mean is {round(cv_results.mean(),5)} and std is {round(cv_results.std(),5)}.")
    print(f"Win10 machine execution time is {execution_time}\n")
```

```python
%%time
# Feature selection using RFECV
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
from sklearn.datasets import make_classification
from sklearn.metrics import classification_report, f1_score, confusion_matrix

# Create the RFECV object using RandomForest classifier best parameters we discovered above, to compute a cross-validated score.
rfc_tuned = RandomForestClassifier(n_estimators = 400, criterion = 'entropy', max_depth = 50, 
                                   random_state =1, n_jobs =-2)

# we will rung RFECV in step size of 3 features
rfecv_step =3
rfecv = RFECV(estimator=rfc_tuned, step=rfecv_step, 
              scoring='f1', cv=StratifiedKFold(3), n_jobs = -2)

rfecv.fit(X_train, y_train)
print(f"Optimal number of features identified by RFECV/RandomForest estimator are : { rfecv.n_features_}")
optimum_number = rfecv.n_features_  # Saving for future use
# below code prints optimum features selected with RFECV
feature_names = dfX_all.columns.tolist()  # list of all features in our dataframe used
mask = rfecv.support_ #list of booleans for selected features
selected_features = [] # The list of our K best features
for bool, feature in zip(mask, feature_names):
        if bool:
            selected_features.append(feature)
            
print(f"{rfecv.n_features_} selected features are : {selected_features}\n")


# Train RandomForest model with selected features on training data and predict for test data
#rfc_tuned = RandomForestClassifier(n_estimators = 400, criterion = 'entropy', max_depth = 50,
#                                    random_state =1, n_jobs =-2)

rfc_tuned.fit(rfecv.transform(X_train), y_train)
y_predicted = rfc_tuned.predict(rfecv.transform(X_test))
score = f1_score(y_test, y_predicted)

   
print(f"Running RandomForest model with test data after rfecv based reduction of features to {rfecv.n_features_}")
print(f"f1 Score  is {score}\n")
rfecv_opt_feature_count = rfecv.n_features_ # saving for future use

# Plot number of features VS. cross-validation scores on training data
fig9, ax9 = plt.subplots(figsize = (7,4))
ax9.set_xlabel("number of features selected")
ax9.set_ylabel("Cross validation score with 3 kfolds, step =5")
ax9.plot(range(0, len((rfecv.grid_scores_) + 1)*rfecv_step, rfecv_step), rfecv.grid_scores_)

ax9.grid(color = 'white')
plt.tight_layout()
plt.show()

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the model
model = Sequential()


# Add input and first hidden layer
number_inputs = X_train.shape[1]
number_hidden_nodes = number_inputs*10
model.add(Dense(units = number_hidden_nodes, activation = 'relu', input_dim=number_inputs))

# # Add second hidden layer
model.add(Dense(units = number_inputs*10, activation = 'relu'))
model.add(Dense(units = number_inputs*10, activation = 'relu'))


#  Add output layer
number_classes = 2
model.add(Dense(units = number_classes, activation = 'softmax'))
model.summary()
```
