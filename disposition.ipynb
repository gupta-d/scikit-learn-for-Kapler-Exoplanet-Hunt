#### Take note that all err1 & err2 related features are uncertainities associated with main feature and therefore no relation to target variable.
#Hence we should either remove these columns, or better we should add average of +ve and -ve uncertainities to the main feature.
#Also this should be done before feature scaling step, as errors are defined on the scale of main feature and should not be independently scaled

```python
df.columns
```

```python
# adding average of +/-ve uncertainity error to main feature
df['koi_period'] = df['koi_period']+(df['koi_period_err1']+df['koi_period_err1'])/2
df['koi_time0bk'] = df['koi_time0bk']+(df['koi_time0bk_err1']+df['koi_time0bk_err2'])/2
df['koi_impact'] = df['koi_impact']+(df['koi_impact_err1']+df['koi_impact_err2'])/2
df['koi_duration'] = df['koi_duration']+(df['koi_duration_err1']+df['koi_duration_err2'])/2
df['koi_depth'] = df['koi_depth']+(df['koi_depth_err1']+df['koi_depth_err2'])/2
df['koi_prad'] = df['koi_prad']+(df['koi_prad_err1']+df['koi_prad_err1'])/2
df['koi_insol'] = df['koi_insol']+(df['koi_insol_err1']+df['koi_insol_err2'])/2
df['koi_steff'] = df['koi_steff']+(df['koi_steff_err1']+df['koi_steff_err2'])/2
df['koi_slogg'] = df['koi_slogg']+(df['koi_slogg_err1']+df['koi_slogg_err2'])/2
df['koi_srad'] = df['koi_srad']+(df['koi_srad_err1']+df['koi_srad_err2'])/2

# now dropping err1 & err2 columns
df = df.drop(['koi_period_err1','koi_period_err2', 
              'koi_time0bk_err1','koi_time0bk_err2', 
              'koi_impact_err1', 'koi_impact_err2',
       'koi_duration_err1', 'koi_duration_err2', 
       'koi_depth_err1', 'koi_depth_err2', 'koi_prad_err1',
       'koi_prad_err2', 'koi_teq', 'koi_insol', 'koi_insol_err1',
       'koi_insol_err2', 'koi_model_snr', 'koi_steff', 'koi_steff_err1',
       'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2',
       'koi_srad', 'koi_srad_err1', 'koi_srad_err2'], axis=1)

```

```python
# Import Modules
from sklearn import model_selection
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier


from datetime import datetime
```

```python
%%time

# feature & target data for training our models
X = X_scaled
y = y_categorical

# classification algorithms to be used (used all default valuse, except a few one where future warnings were issued)
classifiers = [
    SGDClassifier(max_iter=1000, tol=1e-3),
    LinearSVC(dual = False),
    KNeighborsClassifier(),
    SVC(gamma = 'auto'),
    RandomForestClassifier(n_estimators = 100, max_depth = 50),
    MLPClassifier(max_iter = 1000) # default for max-iter is 200, but its failing to converge with defaults value
    ]

models = ['SGD', 'LinearSVC', 'KNN', 'SVC',
         'RandomForest', 'MLP']



# evaluate each model in sequence

results = []
names = []
scoring = 'accuracy'
for model, clf in zip(models, classifiers):
    start_time = datetime.now()
    kf = model_selection.KFold(n_splits=3, random_state=11) # using 3 Kfolds
    cv_results = model_selection.cross_val_score(clf, X, y, cv=kf, scoring=scoring)
    results.append(cv_results)
    names.append(model)
    execution_time = datetime.now()- start_time
    
    print(f"Model {model} cross validation results with 3 folds: {cv_results}.")
    print(f"Mean is {round(cv_results.mean(),5)} and std is {round(cv_results.std(),5)}.")
    print(f"Win10 machine execution time is {execution_time}\n")
    
fig = plt.figure(figsize = (7,4))
ax = fig.add_subplot(111)
ax.set_ylim(bottom =0.4, top =1)
ax.boxplot(results, widths = 0.7)
ax.set_xticklabels(names, rotation = 45)
ax.set_title('CV Scores for ML models')
ax.grid(color = 'blue', linestyle = '--')
ax.set_xlabel("Classifier")
ax.set_ylabel("CV Score with training data")
ax.set_title("Box plot of various model performance using cross-validation")
plt.tight_layout()
# fig.savefig("CV_score_train_ML.png")
plt.show()
```

```python
%%time
# in next step we will try GridSearch for hyperparameter optimization using accuracy_score as metric
#https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

rfc =RandomForestClassifier(random_state =1, n_jobs = -2)
parameters = {'n_estimators' : [200,400,600, 800, 1000], 'criterion' :['gini', 'entropy'], 'max_depth' :[10, 20, 30, 50], 'min_impurity_decrease':[0, 0.1]}
scores = ['accuracy']

for score in scores:
    print(f"# Tuning hyper-parameters for {score}")
    clf = GridSearchCV(rfc, parameters, cv=5, scoring = score)
    clf.fit(X_train, y_train)
    
    print (f'Best score after gridCV training on train dataset: {clf.best_score_}')
    print (f'\nBest parameters set after gridCV on train dataset:')
    best_parameters = clf.best_estimator_.get_params()
    for param_name in sorted(parameters.keys()):
        print (f'{param_name} = {best_parameters[param_name]}')

    y_predict = clf.predict(X_test)
    print (f'\nclassification report on test dataset:\n {classification_report(y_test, y_predict)}')
    print (f'\nconfusion matrix for test dataset: \n {confusion_matrix(y_test, y_predict)}') 
    print(f"Accuracy Score: {round(accuracy_score(y_test, y_predict),5)}\n")
```

#### Grid CV could not help in improving the results. 
Our starting parameters `n_estimators =100, max_depth=50, criterion = 'gini (default)` gives the same accuracy score as GridCV recommendation of `n_estimators =800, max_depth=30, criterion ='entropy`


```python
# lets check if we can reduce the number of features
from sklearn.metrics import (accuracy_score, 
                             confusion_matrix, 
                             classification_report,
                             f1_score, 
                             )
import matplotlib.pyplot as plt

#Keep a list of all features handy for filtering best features based on RFE boolean output mask "rfe.support_" as below
feature_names = dfX_all.columns.tolist() 

from sklearn.feature_selection import RFE
rfc_tuned = RandomForestClassifier(random_state =1, n_estimators =800, max_depth=30, criterion ='entropy')


# #lets check model performance with 10 features onwards
number_features =[]
accuracy_results =[]

for n in range(10,X_train.shape[1]+1): # iterating through all features, dropping sequentially and check results
    rfe = RFE(rfc_tuned, n, step=1)
    rfe = rfe.fit(X_train, y_train)
#     print(rfe.support_)
#     print(rfe.ranking_)

    X_train_rfe = rfe.transform(X_train)
    rfc_tuned.fit(rfe.transform(X_train), y_train)
    X_test_rfe = rfe.transform(X_test)
    y_predict = rfc_tuned.predict(X_test_rfe)
    
    mask = rfe.support_ #list of booleans
    selected_features = [] # The list of our K best features

    for bool, feature in zip(mask, feature_names):
        if bool:
            selected_features.append(feature)
    
    number_features.append(n)
    accuracy_results.append(round(accuracy_score(y_test, y_predict),5))
    
    print(f" Test dataset classification report with top {n} features is \n{classification_report(y_test, y_predict)}")
    print(f" Test dataset confusion_matrix with top {n} features is \n{confusion_matrix(y_test, y_predict)}")
    print(f"{n} selected features are : {selected_features}\n")
    
plt.plot(number_features, accuracy_results)
plt.ylabel('Accuracy Score')
plt.xlabel('# of top features used with Random Forest')
plt.xticks(np.arange(5))
plt.savefig("rfe.png")
plt.show()
    
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, f1_score, confusion_matrix, accuracy_score

knn = KNeighborsClassifier()
parameters = {'n_neighbors':[3,5,10,20,50], 'algorithm':['kd_tree', 'brute', 'auto'], 'leaf_size':[50,100,200]}
scores = ['accuracy']

for score in scores:
    print(f"# Tuning hyper-parameters for {score}")
    clf = GridSearchCV(knn, parameters, cv=5, scoring = score)
    clf.fit(X_train, y_train)
    
    print (f'Best score: {clf.best_score_}')
    print (f'\nBest parameters set:')
    best_parameters = clf.best_estimator_.get_params()
    for param_name in sorted(parameters.keys()):
        print (f'{param_name} = {best_parameters[param_name]}')

    y_predict = clf.predict(X_test)
    print (f'\nclassification report:\n {classification_report(y_test, y_predict)}')
    print (f'\nconfusion_matrix:\n {confusion_matrix(y_test, y_predict)}')
    print(f"Accuracy Score: {round(accuracy_score(y_test, y_predict),5)}\n")
```

```python
%%time
# Lets try if scikit MLP(Neural Network)'s performance could be better
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, f1_score, confusion_matrix, accuracy_score

mlp = MLPClassifier(random_state=1)
parameters = {'activation':['relu'], 'solver' : ['sgd'], 
              'alpha' : [0.0001],
              'learning_rate' : ['adaptive'],
              'hidden_layer_sizes' : [(50,), (50,50),(50,50,50)],
              'max_iter': [1000]}
scores = ['accuracy']

for score in scores:
    print(f"# Tuning hyper-parameters for {score}")
    clf = GridSearchCV(mlp, parameters, cv=5, scoring = score)
    clf.fit(X_train, y_train)
    
    print (f'Best score: {clf.best_score_}')
    print (f'\nBest parameters set:')
    best_parameters = clf.best_estimator_.get_params()
    for param_name in sorted(parameters.keys()):
        print (f'{param_name} = {best_parameters[param_name]}')

    y_predict = clf.predict(X_test)
    print (f'\nclassification report:\n {classification_report(y_test, y_predict)}')
    print (f'\nconfusion_matrix:\n {confusion_matrix(y_test, y_predict)}')
    print(f"Accuracy Score: {round(accuracy_score(y_test, y_predict),5)}\n")
```
